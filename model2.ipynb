```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=190)
```

```python
from sklearn.linear_model import LogisticRegression
m=LogisticRegression(max_iter=1000)
s=m.fit(x_train,y_train)
```

```python
y_pred=s.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.metrics import precision_score
ps=precision_score(y_test,y_pred)
ps
```

```python
from sklearn.metrics import recall_score
rs=recall_score(y_test,y_pred)
rs
```

```python
train_score=m.score(x_train,y_train)
test_score=m.score(x_test,y_test)
print("Training score:",train_score)
print("Testing score:",test_score)
```

SkinThickness: Skin thickness has a weak influence on diabetes with no clear class separation.

```python
from scipy import stats
import statsmodels.api as sm
Glucose_pos=df[df['Outcome']==1]['Glucose']
Glucose_neg=df[df['Outcome']==0]['Glucose']
```

```python
Glucose_pos
```

```python
t_stat, p_val = stats.ttest_ind(Glucose_pos, Glucose_pos)

print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Fail to reject null hypothesis:")
```

#T-Test:


used to compare means when the population variance is unknown or the sample size is small.

#independent T-Test
Example: compare Glucose of positive vs negative person.


Hypothesis:

-H0:Glucose distribution of positive= Glucose distribution of negative



-H1:Glucose distribution of positive != Glucose distribution of negative

```python
from scipy import stats
import statsmodels.api as sm
pregnancies_pos=df[df['Outcome']==1]['Pregnancies']
pregnanacies_neg=df[df['Outcome']==0]['Pregnancies']
```

```python
t_stat, p_val = stats.ttest_ind(pregnancies_pos,pregnanacies_neg)

print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Fail to reject null hypothesis:")
```
